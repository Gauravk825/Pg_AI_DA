```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. Write a Python programming to display a bar chart of the popularity of programming Languages.
data = {"Programming languages": ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'],
'Popularity': [22.2, 17.6, 8.8, 8, 7.7, 6.7]}
df  = pd.DataFrame(data)
df
```

```python
plt.bar(df['Programming languages'], df['Popularity'])
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
```

```python
# 2. Write a Python programming to display a horizontal bar chart of the popularity of programming Languages.

plt.barh(df['Programming languages'], df['Popularity'])
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
```

```python
#Write a Python programming to display a bar chart of the popularity of programming Languages. Use uniform color.
im = plt.imshow(plt.imread(''))
plt.barh(df['Programming languages'], df['Popularity'])
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
```

```python
# Write a Python programming to display a bar chart of the popularity of programming Languages. Use different color for each bar.
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)
#fig, ax = plt.subplots(figsize=(10, 8))
plt.bar(df['Programming languages'], df['Popularity'], color=['blue', 'red', 'pink', 'orange', 'skyblue'])
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.margins(x=0)
plt.xticks(df['Popularity'].index,df['Programming languages'])

plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
# Customize the minor grid
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

```python
# Write a Python program to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.
dist ={
'men' : [22, 30, 35, 35, 26],
'women' : [25, 32, 30, 35, 29]
}

dfmw = pd.DataFrame(dist)
dfmw
```

```python
plt.xlabel("Person")
plt.ylabel("Scores")
plt.title("Score by person")
plt.bar(dfmw['men'], dfmw['women'])
```

```python
plt.figure(figsize=(12,6))
width = 0.3
ind = np.arange(len(dfmw))
plt.bar(ind, dfmw["men"], width=width)
plt.bar(ind+ width, dfmw["women"], width=width)
plt.legend(dfmw.columns)
```
